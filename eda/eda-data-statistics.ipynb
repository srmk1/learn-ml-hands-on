# EDA: Data Statistics and Plots

Once the data from multiple sources have been merged into a single data-set, we look at the overall statistics of the data:
- __Dimesions of the data__:
We look at the dimesions of the data. Usually rows and columns of the data represent the dimensions of the data
    - How many ROWS (also called instances) are there in the data?
    - How many COLS (also called attributes, features) are there in the data?
    - How many Features are there in the data? 
    - Does each Columns represent a feature of the data?

## Attribute Statistics (Univariate)
As a first step, you look at the statistics of individual features one at a time.

- __Varaible/Attribute statistics__:
For each variable/attribute/feature of the dataset, we look at different statistics to understand more about the varaibles. For plain numeric variables, we look at statistics like mean, variance, standard deviation..etc. For catgorical variables we look at histograms, least occuring value, most occuring value kind of statistics. 
    - __Statistics for numeric attributes/variables__
        - Example: mean, variance, median, standard deviation..etc
        - Python panda library df.describe() api gives different statistics for numerical attributes
    - __Statistics for categorical attributes__
        - Example: historgrams, mode, most/least frequent values, percentages, number of unique values
        - Python panda library: Histogram of values: df[attribute].value_counts, panda hist api
        - Seaborn: displot
        - By looking at the histogram of the individual feature, we can look at the overall behaviour of the feature. Is it normally distributed, are there multiple peaks in the data, what does the peak signify
    - __Target statistics__: Targaet variables are your response variable or outcomes
    - __categorical attributes__: For categorical attributes standard statistics like mean, variance, standard deviation does not makes sense. For Categorical attributes, we look at values like - least frequent data, most frequent data, number of unique values, percentiles of values
        - For categorical varaibles, one of the important feature is number of unique values and how many times each unique values occurs in the data. By looking at such data, you can get to know whether your data is balanced or not

## Multivariate Statistics
After you look at each individual features/attribute statistics, we need to look at the relationship between each of the features. Is there any relation between the columns of the data?? You may want to take a look at the correlation between the data.

- __Correlation__: For multi-variable correlation, we can look 
    - Correlation matrix heatmap and Pearson correlation

# Plots 

For each of the variable/feature, we can draw some basic plots
- __Desity Plot__:
- __Histogram__:
- __Box Plot__:
- __Scatter Plot__: Excellent way to find if there is any form of linear relationship between two features/variables.
- __Scatter Plot Matrix__: Is a excellent way to find out any special relation/correlation between different features/variables.
  - It can be used to check if there is a strong linear relationship between the variables. For some of the variable we can derive that there is a pretty strong linear relationship. However for others, we can derive that there is no linear co-relation between these variables. For these kind of variables, how can we quanitfy the linear relation between these variables? For such kind of variables, we may want to use correlation matrix
- __Correlation Matrix__: Can be used to quantify (strong, weak) linear relationship between Numeric variables.
  - Correlation can go as high as 1 or as low as -1
  - Correlation of 1 means two numerical variables/features are perfectly correlated with each other. Its kind of saying x=y
  - Correlation of -1 means like it is saying y=-x
  - Any value b/w 1 and -1 indicates quantification of numerical variables. For example a value of 0 means there is no "linear correlation" between two numerical features/variables. __This does not mean there is no relationship, this just means there is no linear relationship between the two.__ Sometimes looking at number may not be easy, thats when we use heatmap

### Panda Examples

```python
#Sample example
import pandas as pd
from sklearn.datasets import load_breast_cancer

#load breast cancer data from sklearn 
dataset = load_breast_cancer()
#load names for columns
cols = ['V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',
           'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'V29',
           'V30', 'V31', 'V32', 'V33', 'V34', 'V35', 'V36', 'V37', 'V38', 'V39']

df = pd.DataFrame(dataset['data'],columns = cols)
```

```python
#Check how many rows and columns/features are there in dataset
df.shape
```

```python
#show first few rows
df.head()
```

```python
#show datatypes of the columns
df.info()
```

```python
#show summary statistics for each columns
df.describe()
```

```python
#Check target variable properties for ex: V11
df['V11'].value_counts()
```

```python
#Plot the density plot for a variable
df['V11'].plot.kde()
```

```python
#Plot the histogram
df['V11'].plot.hist()
```

```python
#Plot the box plot
df.boxplot(['V11'])
```

```python
#Scatter plot
df.plot.scatter(x='V11',y='V36')
```

```python
#Scatter matrix
#A very efficient way to stop any strong corelation between features/columns
pd.plotting.scatter_matrix(df[['V11','V12','V13']], figsize=(15,15))
```
