```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

#load breast cancer data from sklearn 
dataset = load_breast_cancer()
#load names for columns
cols = ['V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',
           'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'V29',
           'V30', 'V31', 'V32', 'V33', 'V34', 'V35', 'V36', 'V37', 'V38', 'V39']

df = pd.DataFrame(dataset['data'],columns = cols)
```

```python
#Scatter matrix
#A very efficient way to stop any strong corelation between features/columns
pd.plotting.scatter_matrix(df[['V11','V12','V13']], figsize=(15,15))
```

```python
#Basic heatmap for subset of data
import seaborn as sn

#Create a DataFrame using Pandas
df1 = df[['V11','V12','V13']]

#Create a Correlation Matrix using Pandas
corrMatrix = df1.corr()

#Get a Visual Representation of the Correlation Matrix using Seaborn
sn.heatmap(corrMatrix, annot=True, cmap='PiYG')
```

```python
#Basic lineplot on subset of data, without any modification to plot
sn.lineplot(data=df1)
```

```python
#Basic lineplot on subset of data
import matplotlib.pyplot as plt

# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Breast cancer V11, V12, V13 varaibles plot")

# Line chart showing daily global streams of each song 
sn.lineplot(data=df1)
```

```python
# Instead of running plot on all columns you can run only one of column
# Instead of setting data=df1, we set data=df1['V11']. 
# In general, to plot only a single column, we use this format with putting the 
# name of the column in single quotes and enclosing it in square brackets. 
# (To make sure that you correctly specify the name of the column, you can print the 
# list of all column names using the command you learned above.)

#We also add label="V11" to make the line appear in the legend and set its corresponding label.
sn.lineplot(data=df1['V11'], label="V11")
```

```python
#Basic Bar Plot on subset of data, without any modification to plot
sn.barplot(data=df1)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sn.barplot(df1['V11'])

# Add label for vertical axis
plt.ylabel("")
```
