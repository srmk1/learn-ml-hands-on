```python
# import pandas to read data into data-frame
import pandas as pd
df_salary = pd.read_csv('../data-csv/csv-files/Salary_Data.csv')
df_salary['Salary'].tail()
```

```python
#Fitting line equation y=mx+c
import matplotlib.pyplot as plt

x=df_salary['YearsExperience']
y=df_salary['Salary']
plt.scatter(x,y)

# Below loop tries to fit different lines to scatter data
i=1
while i < 10:
    temp=i*1000
    #y=39343+(1000*x)
    y=39343+(temp*df_salary['YearsExperience'])
    plt.plot(x,y)
    i=i+1

# From regressor object we got coeffecient=9312 and intercept=26780
y=26780+(9312*df_salary['YearsExperience'])
plt.plot(x,y)
```

```python
#Fitting line equation y=mx2+c (x-squared)
import matplotlib.pyplot as plt

x=df_salary['YearsExperience']
y=df_salary['Salary']
plt.scatter(x,y)

# Below loop tries to fit different lines to scatter data
i=1
while i < 10:
    temp=i*200
    #y=39343+(1000*x)
    y=39343+(temp*df_salary['YearsExperience']*df_salary['YearsExperience'])
    plt.plot(x,y)
    i=i+1

# From regressor object we got coeffecient=9312 and intercept=26780
y=26780+(9312*df_salary['YearsExperience'])
plt.plot(x,y)
```
