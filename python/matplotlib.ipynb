# Matplotlib
Matplotlib is a Python 2D plotting library which can be used to generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc., with just a few lines of code.

Matplotlib is frequently used in visualising the data as a part of Exploratory Data Analysis stage (of the Machine learning workflow). It is  impossible to manually go through huge data-sets and derive any useful conclusions and hence visualising libraries can be used to derive useful conclusions from the data-sets. One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals.

![](../data-csv/media-pics-videos/anatomy-figure.png)

matplotlib.pyplot is a collection of functions that makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

## Basic Line Plot

```python
import matplotlib.pyplot as plt

# Usually to display plots created by matplotlib,
# we need to plt.show()
# In jupyter notebook, however you can make it inline
# you need not do (matplotlib inline), it will automatically show plot
%matplotlib inline
```

```python
# If you provide a single list or array to the plot() command, 
# matplotlib assumes it is a sequence of y values, and automatically generates the x values for you.
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
#import pandas to read data into data-frame
import pandas as pd
import lxml
```

```python
#URL for wikipedia link
df_cisco_daily = pd.read_csv('../data-csv/csv-files/CSCO.csv')
df_cisco_daily.head()
```

```python
#Lets plot Cisco Share Price movement over last five year
#With Date on x-axis and Share-price on y-axis
#plt.plot(df_cisco_daily['Date'],df_cisco_daily['Close'],'g')

#Experiment with last parameter 'r', 'r-', 'r--', 'r*'
plt.plot(df_cisco_daily['Date'],df_cisco_daily['Close'],'r--')

#Label the x and y-axis
plt.xlabel('Timeline')
plt.ylabel('Share Price')

#Set the title of the plot
plt.title('Cisco 5 Year Share Price History')

#plt.show() not required since we have %matplotlib.inline
```

#### Fixing the x-axis scale

```python
#X-axis scale is not visible in previous plot
#We can use plt.gca (get current axis) to get axis
#Then we use set_major_locator to have only 5-entries on plot

#import LinearLocator
from matplotlib.ticker import LinearLocator

plt.plot(df_cisco_daily['Date'],df_cisco_daily['Close'],'r--')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=6))
```

#### Change size of the plot

```python
#Lets change the size of the plot
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(df_cisco_daily['Date'],df_cisco_daily['Close'],'r--')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=15))
```

## Multi-Line Plot
Allows to plot multiple variable on same x and y-axis

```python
#Multi-line plot
#Set the size of the plot to 10x5
plt.rcParams["figure.figsize"] = (10,5)

#Select first 30 entries from the data
df_thirty_day = df_cisco_daily.iloc[1:30]

#Plot the columns high and date in red
plt.plot(df_thirty_day['Date'],df_thirty_day['High'], 'r')
#Plot the columns Low and date in red
plt.plot(df_thirty_day['Date'],df_thirty_day['Low'], 'b')


#Set the scale of x-axis to have only 15 values
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=5))
```

## Subplots
The Matplotlib subplot() function can be called to plot two or more plots in one figure. Matplotlib supports all kind of subplots including 2x1 vertical, 2x1 horizontal or a 2x2 grid.

```python
#Creating subplots
#subplot(ncols, nrows, index)

#Create a plot at index 1
plt.subplot(2,2,1)
plt.plot(df_thirty_day['Date'],df_thirty_day['High'],'b')
plt.title('Day High')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=5))

#Create a plot at index 2
plt.subplot(2,2,2)
plt.plot(df_thirty_day['Date'],df_thirty_day['Low'],'r*')
plt.title('Day Low')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=5))

#Create a plot at index 3
plt.subplot(2,2,3)
plt.plot(df_thirty_day['Date'],df_thirty_day['Open'],'g*-')
plt.title('Day Open')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=5))

#Create a plot at index 4
plt.subplot(2,2,4)
plt.plot(df_thirty_day['Date'],df_thirty_day['Close'],'y--')
plt.title('Day Close')
ax = plt.gca()
ax.get_xaxis().set_major_locator(LinearLocator(numticks=5))
```

## Histograms
A histogram is an approximate representation of the distribution of numerical or categorical data. 
* To construct a histogram, the first step is to "bin" (or "bucket") the range of valuesthat is, divide the entire range of values into a series of intervalsand then count how many values fall into each interval.
* The bins are usually specified as consecutive, non-overlapping intervals of a variable. The bins (intervals) must be adjacent, and are often (but not required to be) of equal size

### Interpreting Histograms
The words used to describe/interpret the patterns in a histogram are: "symmetric", "skewed left" or "right", "unimodal", "bimodal" or "multimodal".

![](../data-csv/media-pics-videos/interpret-histogram.png)

```python
#How about creating a histogram out of it
#All the prices will be put into 10bins and counted. 
#By default prices will be categorised into 10bins (which will give density of 10 ranges)
#If you need more you need to send "bins=20" parameter
#Count(density) will be on y-axis
#Price range will be in x-axis
plt.hist(df_cisco_daily['Close'])
plt.xticks(rotation=90)
plt.title("Histogram of 5 year Cisco Share Price History")
```

```python
#Lets set bins=30, so that share price is catgorised into 30 different price-ranges
plt.hist(df_cisco_daily['Close'],bins=30)
plt.xticks(rotation=90)
plt.title("Histogram of 5 year Cisco Share Price History")

#You can observe that there around 120+ days in which share price was 32$ and 28$ (totally 240+ days)
#There are only 10days in share price has gone less than 10$
```

## Box Plot
Box plots divide the data into sections that each contain approximately 25% of the data in that set.

![](../data-csv/media-pics-videos/5-number-repesent.png)

Box plots are useful as they provide a visual summary of the data enabling researchers to quickly identify mean values, the dispersion of the data set, outliers and signs of skewness.

### Interpreting Box Plot

![](../data-csv/media-pics-videos/box-plot.png)

* Minimum Score: The lowest score, excluding outliers (shown at the end of the left whisker).
* Lower Quartile: Twenty-five percent of scores fall below the lower quartile value (also known as the first quartile).
* Median: The median marks the mid-point of the data and is shown by the line that divides the box into two parts (sometimes known as the second quartile). Half the scores are greater than or equal to this value and half are less.
* Upper Quartile: Seventy-five percent of the scores fall below the upper quartile value (also known as the third quartile). Thus, 25% of data are above this value.
* Maximum Score: The highest score, excluding outliers (shown at the end of the right whisker).
* Whiskers: The upper and lower whiskers represent scores outside the middle 50% (i.e. the lower 25% of scores and the upper 25% of scores).
* The Interquartile Range (or IQR): This is the box plot showing the middle 50% of scores (i.e., the range between the 25th and 75th percentile).

```python
#How to draw Box Plot?
plt.subplot(2,2,1)
plt.boxplot(df_cisco_daily['Close'])

#vert=Flase - will draw horizontal box plot
plt.subplot(2,2,2)
plt.boxplot(df_cisco_daily['Open'], vert=False)

#patch_artist=True - will draw solid boxes
plt.subplot(2,2,3)
plt.boxplot(df_cisco_daily['High'], patch_artist=True)

#Box plots also show the outliers
plt.subplot(2,2,4)
plt.boxplot(df_cisco_daily['Volume'])
```
